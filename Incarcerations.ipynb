```python
pris = pd.read_csv('ICPSR_37003/DS0001/37003-0001-Data.tsv', delimiter='\t')
pris
```

```python
cols = ['YEAR','STATE']+[col for col in pris.columns if 'TOT' in col]
pris[cols].head()
```

```python
pris.columns
```

```python
for col in ['PVINCLM', 'PVINCLF']:
    print(col, pris[col].value_counts(normalize=True))
```

```python
jurisdictions = ['JURTOTM', 'JURTOTF']
privmales = ['PVINM', 'PVOTHM']
privfemales = ['PVINF', 'PVOTHF']
pris['PVM'] = pris[privmales].sum(axis=1)
pris['PVF'] = pris[privfemales].sum(axis=1)
pris['PercPrivM'] = pris['PVM'] / pris['JURTOTM']
pris['PercPrivF'] = pris['PVF'] / pris['JURTOTF']
pris[['PercPrivM','PercPrivF']].info()
```

```python
pris[privmales].head()
```

```python
pris[pris['PVINM']>0].groupby('STATE')['PVINM'].agg(['min', 'max', 'mean']).head()
```

```python
PVMSTYR = pris[(pris['PVINM']>=0)][['YEAR','PVINM','STATE']].pivot(index='YEAR', columns='STATE', values='PVINM')
PVFSTYR = pris[(pris['PVINF']>=0)][['YEAR','PVINF','STATE']].pivot(index='YEAR', columns='STATE', values='PVINF')
PVMSTYR
```

## Florida Prisons Are Becoming More and More Privatized

```python
sns.set_style('darkgrid')
```

```python
plt.figure(figsize=(10,6))
PVMSTYR['FL'].plot()
PVFSTYR['FL'].plot();
```

```python
PVMPerSTYR = pris[(pris['PVINM']>=0)][['YEAR','PercPrivM','STATE']].pivot(index='YEAR', columns='STATE', values='PercPrivM')
PVFPerSTYR = pris[(pris['PVINF']>=0)][['YEAR','PercPrivF','STATE']].pivot(index='YEAR', columns='STATE', values='PercPrivF')
```

```python
plt.figure(figsize=(10,6))
PVMPerSTYR['FL'].plot()
PVFPerSTYR['FL'].plot();
```

```python
PVMPerSTYR.mean(axis=1).plot()
```
